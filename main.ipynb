# 4. Model Building

### Unsupervised Machine Learning

- Elbow Method

```python
from sklearn import cluster
error = []
for i in range(1,10):
    kmeans = cluster.KMeans(n_clusters= i, init= 'k-means++', random_state=0)
    kmeans.fit(KMeansdf)
    error.append(kmeans.inertia_)
error
```

```python
plt.plot(range(1, 10), error)
plt.title("Elbow Method")
plt.xlabel("No of Cluster")
plt.ylabel("Error")
plt.show()
```

- Intialize Model

```python
km = cluster.KMeans(n_clusters= 4)
y_pred = km.fit_predict(KMeansdf)
```

- Dimensionality Reduction 

```python
from sklearn.decomposition import PCA
pca = PCA(n_components= 2)
dfPCA = pca.fit_transform(KMeansdf)
dfPCA = pd.DataFrame(dfPCA, columns= ['PCA1', 'PCA2'])
dfPCA['Cluster'] = y_pred
dfPCA.head()
```

# ERROR IN PLOTTING THE CLUSTER NEED TO FIGURE OUT WHATS HAPPENING 

```python
# Lets plot the cluster
plt.figure(figsize= (10, 10))
sns.scatterplot(data=dfPCA, x='PCA1', y='PCA2', hue='Cluster')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color= 'black', s=300, marker='*', label= 'centroid')
plt.legend()
```

# Supervisied Machine Learning 

- Splitting Dataset

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns= 'Revenue')
y = df.Revenue

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=  0.2, random_state=42)
```

```python
# Lets define a evaluator that does metrics 

def evaluater(model, x_test, y_test):
    from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
    y_pred = model.predict(x_test)

    print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
    print(f"Confusion Matrix:\n {confusion_matrix(y_test, y_pred)}")
    print(f"Classification Report: \n {classification_report(y_test, y_pred)}")
```

- Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
evaluater(lr, x_test,  y_test)
```

- Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
evaluater(dt, x_test, y_test)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train, y_train)
evaluater(svc, x_test, y_test)
```

- Guassian Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
evaluater(gnb, x_test, y_test)
```

- K Nearest Neighbours Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
evaluater(knn, x_test, y_test)
```
